## Installation of the free Maestro Viewer

 1. Navigate to [`schrodinger.com`](https://www.schrodinger.com)
 2. Click on `LOGIN` on the top right
 3. Select `create an account for free` and follow the instructions to create a free account (or enter your account e-mail if you already have an account).
 4. Once your account is active, login and search for `Maestro Viewer` using the search button next to `My Account`. The first search item should be the `Maestro Viewer` link. Click on it and select `Download`, select your laptop's operating system (Windows, Mac, Linux) and proceed to download the software.
 5. While the Maestro Viewer is downloading (it's several GB, be patient) study the installation instructions for your operating system. Disregard the part about obtaining a license: it's included in the installer.
 6. Follow the installation instructions and try starting the Maestro Viewer once the software download is complete.
 7. As a convenience, the Maestro Viewer installers are also available at this [Dropbox link](https://www.dropbox.com/scl/fo/znlw50pi5i5ej0sw8l62w/ALHlxl5OerjQ-X206dzQJG0?rlkey=fxnbnum5ndpf181z4uhilqryo&st=v16o2hic&dl=0). However, do not share this link outside class, we do not have permission to redistribute Schrodinger's software.


## Google Colab PRO account

1. Assuming you already have a Google (GMail) account: navigate to [`colab.google.com`](https://colab.research.google.com/) and hit `Cancel` if a popup window appears.
2. Open Settings on the top left
3. Select `Colab Pro` in the menu on the right.
4. Follow the instructions to obtain a free Colab PRO subscription for faculty and students. It requires your official campus e-mail.
